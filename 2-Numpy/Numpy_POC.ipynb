# **Numpy**
The NumPy library is the core library for scientific computing in
 Python.
It provides a high-performance multidimensional array
 object, and tools for
working with these arrays.


```python
#Use the following import convention:
import numpy as np
```

## Creating Arrays


```python
#One Dimentional numpy array
a = np.array([1,2,3])
print(a)
#Two Dimentional numpy array
b = np.array([(1.5,2,3), (4,5,6)])
print(b)
#Three Dimentional numpy array
c = np.array([[(1.5,2,3), (4,5,6)],[(3,2,1), (4,5,6)]], dtype = float) # Specifying datatype is not necessary
print(c)
```

## Initial Placeholders

```python
#Create an array of zeros
print(np.zeros((3,4)))
#Create an array of ones
print(np.ones((2,3,4),dtype=np.int16))
#Create an array of evenly spaced values (number of samples)
print(np.linspace(0,2,9))
```

```python
#Create an array with random values
print(np.random.random((2,2)))
#Create an empty array
print(print(np.empty((3,2))))
#Create an array of evenly spaced values (step value)
d = np.arange(10,25,5)
print(d)
```

```python
#Create a constant array
e = np.full((2,2),7)
print(e)
#Create a 2X2 identity matrix
f = np.eye(2)
print(f)
```

## **Array Mathematics**
## Arithmetic Operations

```python
#Addition using operator
print(b + a)
#Addition using NumPy class method
np.add(b,a)
```

```python
#Subraction using operator
g = a - b
print(g)
#Subraction using NumPy class method
np.subtract(a,b)
```

```python
#Multiply using operator
print(a * b)
#multiply using NumPy class method
np.multiply(a, b)
```

```python
#Divide using operator
print(a / b)
#Divide using NumPy class method
np.divide(a, b)
```

```python
#Exponentiation
print(np.exp(b))
#Square root
print(np.sqrt(b))
#Print sines of an array
print(np.sin(a))
#Element-wise cosine
print(np.cos(b))
#Element-wise natural logarithm
print(np.log(a))
#Dot product
print(e.dot(f))
```

## Comparison

```python
#Element-wise comparison
a == b
```

```python
#Element-wise comparison
a < 2
```

```python
#Array-wise comparison
np.array_equal(a, b)
```

## Aggregate Functions

```python
#Array-wise  sum
print(a.sum())
#Array-wise  minimum  value
print(a.min())
#Maximum  value  of  an  array  row
print(b.max(axis=0))
#Cumulative sum of the elements
print(b.cumsum(axis=1))
#Mean
print(a.mean())
#Median
print(np.median(b))
#Correlation  coefficient
print(np.corrcoef(a))
#Standard  deviation
print(np.std(b))
```

## **Copying Arrays**

```python
#Create a view of the array with the same data
h = a.view()
print(h)
#Create a copy of the array
print(np.copy(a))
#Create a deep copy of the array
h = a.copy()
print(h)
```

## **Sorting Arrays**

```python
#Sort an array
a.sort()
#Sort the elements of an array's axis
c.sort(axis=0)
```

## **Subsetting, Slicing, Indexing**
## Subsetting

```python
#Select the element at the 2nd index
print(a[2])
#Select the element at row 1 column 2 (equivalent to b[1][2])
print(b[1,2])
```

## Slicing

```python
#Select items at index 0 and 1
print(a[0:2])
#Select items at rows  0 and 1 in column 1
print(b[0:2,1])
#Select all items at row  0 (equivalent to b[0:1, :])
print(b[:1])
#Same as [1,:,:]
print(c[1,...])
#Reversed array a array([3, 2, 1])
print(a[ : :-1])
```

## Boolean Indexing

```python
#Select elements from a less than 2
print(a[a<2])
```

## Fancy Indexing

```python
#Select elements (1,0),(0,1),(1,2) and (0,0)
print(b[[1,  0,  1,  0],[0,  1, 2, 0]])
#Select a subset of the matrix’s rows  and columns
print(b[[1,  0,  1,  0]][:,[0,1,2,0]])
```

## **Array Manipulation**
## Transposing Array

```python
#Permute array dimensions
i = np.transpose(b)
print(i)
#Permute array dimensions
i.T
```

## Changing Array Shape

```python
#Flatten the array
print(b.ravel())
#Reshape, but don’t change data
print(g.reshape(3,-2))
```

## Adding/Removing Elements

```python
#Return a new array with shape (2,6)
h.resize((3,1))
print(h)
#Append items to an array
print(np.append(h,g))
#Insert items in an array
print(np.insert(a, 1, 5))
#Delete items from an array
print(np.delete(a,[1]))
```

## Combining Arrays

```python
#Concatenate arrays
print(np.concatenate((a,d),axis=0))
#Stack arrays vertically (row-wise)
print(np.vstack((a,b)))
#Stack arrays vertically (row-wise)
print(np.r_[e,f])
#Stack arrays horizontally (column-wise)
print(np.hstack((e,f)))
#Create stacked column-wise arrays
print(np.column_stack((a,d)))
#Create stacked column-wise arrays
print(np.c_[a,d])
```

## Splitting Arrays

```python
#Split the array horizontally at the 3rd  index
print(np.hsplit(a,3))
#Split the array vertically at the 2nd index
np.vsplit(c,2)
```
